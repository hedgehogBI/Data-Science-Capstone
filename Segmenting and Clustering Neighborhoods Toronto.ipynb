Import all required libraries 

```python
import pandas as pd
import numpy as np 
import bs4 as bs
import urllib.request
```

Manage the extract by utilising Beautiful soup

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

response=requests.get(url)

html = response.text

page = bs.BeautifulSoup(html,'html.parser')

print(page)
```

Extract the data in the table

```python
Toronto = page.find('table',{'class':'wikitable sortable'})
Toronto
```

identify all rows to build the dataframe

```python
rows=list()
for row in Toronto.findAll("tr"):
   rows.append(row)
rows
```

Convert the data into a dataframe 

```python
toronto_df = []
for tr in rows:
    td = tr.find_all('td')
    row = [tr.text for tr in td]
    toronto_df.append(row)
df=pd.DataFrame(toronto_df, columns=["Postal Code", "Bourough", 'Neighborhood'])
df
```

```python
df.head()
```

Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.
Step 1 replace the Not assigned with nan for selected column Bourough	
Step 2 Drop nan value 

```python
df_cleaned = df.replace('Not assigned\n',np.nan, regex=True)
df_cleaned.head()
df_cleaned.dropna(subset=["Bourough"], axis=0, inplace=True)
df_cleaned = df_cleaned.reset_index(drop=True)
df_cleaned.head(10)
```
